## Import Module

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
import warnings
%matplotlib inline

plt.style.use('seaborn')  # seanborn scheme 사용
sns.set(font_scale=2.5)  #  set seaborn graph font size in 2.5
warnings.filterwarnings('ignore')  # warning 무시
```

## Process
1. **데이터셋 확인** - 대부분 케글 데이터는 잘 정제되어 있다. 하지만 null data가 존재한다. 이를 확인하고 향후 수정한다.
2. **탐색적 데이터 분석** (EDA, Exploratory Data Analysis) - 여러 feature들을 개별적으로 분석하고, feature들 간의 **상관관계**를 확인. 여러 시각화 툴을 사용해 insight를 얻는다.
3. **feature engineering** - 모델을 세우기 앞서, 모델의 성능을 높일 수 있도록 feature들을 engineering한다. One-hot encoding, class 나누기, 구간으로 나누기, 텍스트 데이터 처리 등을 한다. 
4. **Model 만들기** - sklearn을 사용해 모델을 만든다. Python에서 ML을 할때 sklearn을 사용하면 수많은 알고리즘을 일관된 문법으로 사용할 수 있다. 물론 딥러닝을 위한 Framework인 TF, Pytorch를 사용할 수도 있다.
5. **모델 학습 및 예측** - trainset을 가지고 모델을 학습 시킨 후, testset을 가지고 prediction을 진행
6. **모델 평가** - 예측 성능이 원하는 수준인지 판단한다. 풀려는 문제에 따라 모델을 평가하는 방식도 달라진다. 학습된 모델이 어떤 것을 학습했는지 확인해본다.

나이대가 변함녀서 생존률이 어떻게 되는지 보자.

나이 범위를 넓혀가며, 생존률이 어떻게 되는지 보자.

```python
cummulate_survival_ration = []
for i in range(1, 80): 
    cummulate_survival_ration.append(df_train[df_train['Age'] < i]['Survived'].sum() / 
                                     len(df_train[df_train['Age'] < i]['Survived']))
    
plt.figure(figsize=(7, 7))
plt.plot(cummulate_survival_ration)
plt.title('Survival rate change depending on range of Age', y=1.02)
plt.ylabel('Survival rate')
plt.xlabel('Range of Age(0~x)')
plt.show()
```

- 어릴수록 생존률이 높다.
- 나이가 중요한 feature로 사용될 수 있다.

### 2.8 Fare

- Fare는 탑승요금이고 continous feature이다. 

```python
f, ax = plt.subplots(1, 1, figsize=(8, 8))
g = sns.distplot(df_train['Fare'], color='b', label='Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```

- Distribution은 매우 비대칭이다(high skewness). 만약 이것을 그대로 모델에 넣어준다면 모델이 잘못 학습할 수 있다. 몇개 없는 outlier에 대해서 너무 민감하게 반응한다면, 실제 예측 시에 좋지 못한 결과를 부를 수 있다.

- outlier의 영향을 줄이기 위해 Fare에 log를 취하자.

- DataFrame의 특정 columns에 공통된 작업(함수)를 적용하고 싶으면 map, apply 를 사용하면 손쉽게 적용 가능하다.

- Fare column에 log를 취하는 것인데, Python의 Lambda 함수르 이용해 간단한 로그를 적용하는 함수를 map에 인수로 넣어주면 Fare columns 데이터에 그대로 적용된다. 

```python
df_test.loc[df_test.Fare.isnull(), 'Fare'] = df_test['Fare'].mean()

df_train['Fare'] = df_train['Fare'].map(lambda i: np.log(i) if i > 0 else 0)
df_test['Fare'] = df_test['Fare'].map(lambda i: np.log(i) if i > 0 else 0)

f, ax = plt.subplots(1, 1, figsize=(8, 8))
g = sns.distplot(df_train['Fare'], color='b', label='Skewness: {:.2f}'.format(df_train['Fare'].skew()), ax=ax)
g = g.legend(loc='best')
```

- log를 취하니, 비대칭성이 많이 사라진 것을 볼 수 있다.
- 이런 작업을 사용해 모델이 좀 더 좋은 성능을 내도록할 수 있다.
- 이 작업은 feature engineering에 들어가는 부분이다.
- 모델을 학습시키기 위해, 그 모델의 성능을 높이기 위해 feature들에 조작을 가하거나, 새롱누 feature를 추가하는 것을 feature engineering이라고 한다.
