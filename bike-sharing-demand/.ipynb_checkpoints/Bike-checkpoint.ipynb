## Preprocessing

```python
train_casual = train_df.casual
train_registered = train_df.registered

train_copy = train_df.copy()

train_copy.drop(['casual', 'registered'], axis=1, inplace=True)

train_index = train_df.shape[0]
test_index = test_df.shape[0]

all_df = pd.concat((train_copy, test_df)).reset_index()

del all_df["index"]
```

```python
all_df.head()
```

```python
pre_df = all_df.copy()

# change all_df['datetime'] type to datetime
pre_df['datetime'] = pd.to_datetime(pre_df['datetime'])

# Split the datetime into year, month, day, time
year = pre_df["datetime"].dt.year
month = pre_df["datetime"].dt.month
day = pre_df["datetime"].dt.day
time = pre_df["datetime"].dt.time
weekday = pre_df["datetime"].dt.weekday

day_li = ["Mon", "Tue", "Wen", "Thur", "Fri", "Sat", "Sun"]

pre_df = pre_df.merge(pd.get_dummies(pre_df["datetime"].dt.weekday.apply(lambda x: day_li[x]), prefix=None), left_index=True, right_index=True)

pre_df
```

### Season
- 1 = spring
- 2 = summer 
- 3 = fall 
- 4 = winter

2011-01-01 == 1, Spring $\rightarrow$ Winter

$\Rightarrow$ One-hot-encoding

### Weather

- 1: Clear, Few clouds, Partly cloudy, Partly cloudy
- 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist
- 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds
- 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog 

$\Rightarrow$ One-hot-encoding

```python
def changeSeason(month):
    if month in [12, 1, 2]:
        return 4
    elif month in [3, 4, 5]:
        return 1
    elif month in [6, 7, 8]:
        return 2
    elif month in [9, 10, 11]:
        return 3

pre_df.season = pre_df.season.apply(changeSeason)
```

```python
pre_df = pre_df.merge(pd.get_dummies(pre_df.season, prefix="season"), left_index=True, right_index=True)
pre_df = pre_df.merge(pd.get_dummies(pre_df.weather, prefix="weather"), left_index=True, right_index=True)
pre_df.head()
```

## Evaluation

RMSLE(Root Mean Squared Logarithmic Error): $\sqrt{\frac{1}{n}\sum^{n}_{i=1}(log(p_i + 1) - log(a_i + 1))^2}$

```python
def rmsle(y, y_):
    log1 = np.nan_to_num(np.log(y + 1))
    log2 = np.nan_to_num(np.log(y_ + 1))
    cal = (log1 - log2) ** 2
    return np.sqrt(np.mean(cal))
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
lr = LinearRegression()
y_log = np.log1p(yLabels)
```

```python
lr.fit(train, y_log)

preds = lr.predict(train)

print('RMSLE Value for Linear Regression: {}'.format(rmsle(np.exp(y_log), np.exp(preds))))
```
